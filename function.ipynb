```python
#Write a function that takes a number and returns the reverse of the number.
def  return_num(n):
   rev=0
   while n>0:
    rem=n%10
    rev=rev*10+rem
    n//=10
   return rev
return_num(123)

```

```python
# 10 Write a function that takes a number and returns whether it is a palindrome or not.
def number(n):
   rev=0
   temp=n
   while temp>0:
    rem=temp%10
    rev=rev*10+rem
    temp//=10
   if rev==n:
    return "palindrom"
   else:
    return "not"
number(121)
```

```python
# Write a function that takes a number and returns whether it is a prime number or not.
def prime(n):
  count=0
  for i in range(1,n+1):
    if n%i==0:
      count+=1
  if count==2:
    return "prime"
  else:
    return "no"
prime(8)
```
